# __인공지능 사례 분석__
> 언어, 음성, 이미지, 자율주행 분야(4가지)에서 인공지능을 활용한 제품 및 서비스를 알아본다.



## __1. 언어__


### __1.1. 인공신경망 기계번역(NMT, Neural Machine Translation)__
- 원문 번역의 전 과정을 인공 신경망으로 처리하는 기계 번역으로, 주로 심층 신경망(DNN, Deep Neural Network)을 사용

#### __1.1.1. Amazon Translate__
   - 소셜 미디어 콘텐츠의 다국어 감성 분석 지원
   - 사용자 생성 콘텐츠에 대한 온디맨드 번역 제공
   - 커뮤니케이션 어플리케이션에 실시간 번역 가능

#### __1.1.2. 구글 번역기__
   - 웹에서 번역 콘텐츠를 긁어와서 학습하기 때문에 인터넷 공간에서 만들어진 언어도 번역 가능
   - 데이터가 부족한 언어의 학습 가능
   - 성별을 특정하지 않는 대명사를 번역할 땐 인간의 편견을 반영한 결과물을 내뱉기도 함


### __1.2. 자연어 처리(NLP, Natural language processing)__
- 컴퓨터를 이용하여 인간 언어의 이해, 생성 및 분석을 다루는 인공 지능 기술

#### __1.2.1. AIQ.토크 챗봇__
   - 딥러닝 기반 방식과 자연어 처리(NLP) 기법을 결합한 하이브리드 방식으로 설계된 챗봇 빌더 플랫폼
   - 버전 관리 및 배포 시스템, 운영 현황 대시보드 등 다수 관리자가 효율적으로 챗봇을 운영할 수 있게 하는 기능 지원
   - 음성 기술과 결합해 업무 자동화, 비대면 비즈니스 등 다양한 분야로의 비즈니스 확장도 모색할 계획



## __2. 음성__


#### __2.1. 애플이 인수한 보이시스(Voysis)__
   - 온라인 쇼핑앱들에 탑재된 디지털 음성 비서가 사람들의 음성 명령을 보다 정확하게 반응할 수 있도록 하는 기술에 초점
   - 2018년 AI가 일단 한번 훈련되면 적게는 25메가바이트(MB) 메모리를 사용하는 수준으로 자사 소프트웨어 시스템을 축소했다고 함

#### __2.2. 아마존 '알렉사'__
   - 머신러닝으로 수많은 뉴스 오디오 클립으로부터 낭독 기술을 배운 뒤 자체적으로 뉴스를 읽어줌.
   - 뉴스 진행자가 읽어주는 것 같은 뉴스 낭독 품질을 구현하기 위해 문자를 음성으로 바꿔주는 기술인 TTS(Text-to-speech) 기술에 머신러닝을 결합

#### __2.3. 네이버, 새 음성인식 기술 ‘NEST'__
   - 제한된 데이터 학습으로도 복잡하고 다양한 장문의 음성 표현을 정확하게 인식하고 텍스트로 변환할 수 있는 기술
   - 학습에 필요한 데이터의 양과 시간을 기존의 10분의1 수준으로 단축시키면서 인식의 정확도는 오히려 높였다는 설명
   - AI가 전화로 코로나19의 능동감시자를 확인하는 ‘클로바 케어콜’ 서비스에 적용



## __3. 이미지__


#### __3.1. 5G 스마트팩토리 비전__
   - 공장에 설치된 카메라들을 통해 이미지를 수집하고 데이터를 분석해 불량검사, 제품식별, 치수측정 등 기존에 사람이 육안으로 하던 검사 작업을 인공지능으로 활용하는 서비스
   - 정해진 규칙에 따라 검사작업을 진행하는 ‘룰(Rule) 기반의 머신비전’ 솔루션과, 비정형 항목의 검사가 가능한 ‘딥러닝(Deep Learning) 기반의 머신비전’ 솔루션으로 구성

#### __3.2. CheXNeXt__
   - 촬영한 부위가 어디인지 알아내는 것뿐 아니라 폐렴(Pneumonia)과 기흉(Pneumothorax), 무기폐(Atelectasis), 심장 비대(Cardiomegaly), 흉수(Effusion), 폐기종(Emphysema), 폐섬유화증(Fibrosis) 등 14가지 병리에 대해서 전문적으로 진단할 수 있도록 훈련받음
   - 약 90초만에 모든 이미지를 판독해 질환을 진단
   - 방사선과 전문의의 도움 없이 X선 이미지를 판독해 질병을 정확하게 진단하는 것이 목표



## __4. 자율주행__


### __4.1. 레이더(RADAR) 센서__
- 전자파 기반 센서로 전자파를 발사해 반사돼 돌아오는 신호를 기반으로 주변 사물과의 거리, 속도, 방향 등의 정보를 추출
- 레이더는 물체의 형상을 확실하게 인식할 수는 없지만, 날씨나 시간과 관계없이 제 성능을 발휘하는 가장 신뢰도가 높은 센서 
- 레이더는 주파수에 따라 단거리부터 중거리, 장거리를 모두 감지 할 수 있어 현재도 긴급자동제동장치, 스마트 크루즈 컨트롤 등
다양한 ADAS기술에 적용되고 있음

#### __4.1.1. 자율주행 로봇 공기청정기 ATMOBOT Pro__
   - 최첨단 레이저 거리 센서(LDS)로 최대 200msq의 실내공간을 스캔하고, SLAM기술(실내 GPS기술)을 통해 정교한 가상 지도를 만들어 냄
   - 이동 정화 모드에서는 LDS 레이저 거리 센서와 40여 세트의 추가 센서를 통해 집 안 구조와 각종 가구 등의 위치를 파악
   - 상부의 적외선 센서는 높이가 높은 장애물을 파악하며, 장애물 발견 시 충돌하지 않도록 감속하거나 다른 경로로 우회하며, 하부의 초음파 센서는 높이가 낮거나 바닥에 위치한 장애물을 파악해 공기청정기의 안전한 주행을 도움


### __4.2. 라이다(LiDAR) 센서__
- 특정 신호를 보내고, 반사되는 신호를 다시 받아
대상을 추정한다는 점에서 레이더와 닮았지만,
전파 대신 직진성이 강한 고출력 펄스 레이저로 고정밀 데이터를 확보
- 폭과 거리, 높낮이까지 반영한 3차원의 점을 한데 모아 사물의 형상 데이터를 추출하는 것
- 특히 라이다 센서는 고해상도의 3차원 공간 정보를 획득할 수 있으며, 오차범위가 CM단위에 불과할 정도로 정확도가 굉장히 뛰어남


### __4.3. 무인 자율주행 전동 킥보드 '고엑스'(Go X)__
   - 사용자가 앱을 이용해 전동 킥보드를 호출하면 인근에 있는 고엑스 킥보드 스스로 이동해 사용자가 있는 위치로 이동
   - 인공지능과 위치추적 기능, 간단한 보조바퀴를 이용해 킥보드 자율주행 기능을 구현
   - 승용차처럼 복잡한 자율주행 기능 구현이 없이도 충분히 서비스에 적용할 수 있을지를 테스트하는 중
